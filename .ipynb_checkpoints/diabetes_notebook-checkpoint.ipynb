## ML MODELS

```python
dataset_ml = dataset_eda.copy()
```

```python
dataset_ml['Outcome'].value_counts()
# here dominant class is 0 so to balance that we will use smote 
```

```python
dataset_ml.head(3)
```

```python
x  = dataset_ml.drop('Outcome',axis=1)
```

```python
x.head() 
```

```python
y = dataset_ml['Outcome']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3, random_state=0)
# spllitong the dataset before smote to avoid dataleakage
```

```python
# now use smote fit_resample (xtran,ytrain) to balance the dominant class of outcome column 
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
x_train_resampled , y_train_resampled = smote.fit_resample(x_train ,y_train )
```

```python
print(x_train_resampled.shape)
print(x_train.shape)
```

```python
print(y_train_resampled.shape)
print(y_train.shape)
# clearly there are addon samples
```

```python
print(y_train_resampled.value_counts())
# clearly balanced classes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
models = {
    'lr' :LogisticRegression(),
    'Rf' : RandomForestClassifier(n_estimators=5),
    'knn': KNeighborsClassifier(n_neighbors=5),
    'svm' : SVC(kernel='linear')
}
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score ,classification_report ,confusion_matrix
```

```python
sc = StandardScaler()
```

```python
x_train_resampled_scaled =  sc.fit_transform(x_train_resampled)
```

```python
x_test_scaled = sc.transform(x_test)
```

```python
models_acc = {}
```

```python
for name , model in models.items():
    model.fit(x_train_resampled_scaled , y_train_resampled)
    y_models_predict = model.predict(x_test_scaled)
    acc_models = accuracy_score(y_models_predict , y_test)*100
    print(f'{name} {'\n'} {classification_report(y_models_predict , y_test)}')
    models_acc.update({name : acc_models})
```

```python
print(models_acc)
```

```python
sns.barplot(x = list(models_acc.keys()) , y = list(models_acc.values()))
plt.title('Model  vs Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')
# lr has highest accuracy with 75.75 %
#then knn and Rf has same accuracy of 74.02 %
# then svm has accuracy of 73.59 %
```

### Optimised LR model using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'solver' : ['liblinear' , 'newton-cholesky' ,'sag', 'saga' , 'lbfgs' ],
    # liblinear and newtom-cholesky are used for binary
    # , 'sag', 'saga' , 'lbfgs'  as multiclasses 
    
    'C' :[0.01 ,0.1 ,0.5, 1 ,10] ,
    # regularization strenght
    
    'penalty' : ['l2']
    # for newton-cholesky algo
    
}
```

```python
lr_gcv_model =  GridSearchCV(estimator= LogisticRegression()  , scoring='f1',param_grid= param_grid , cv=5 ,n_jobs=-1 ,verbose=2)
```

```python
lr_gcv_model.fit(x_train_resampled_scaled , y_train_resampled)
```

```python
lr_gcv_y_predict = lr_gcv_model.predict(x_test_scaled)
```

```python
print(lr_gcv_model.best_estimator_)
```

```python
print(lr_gcv_model.best_params_)
```

```python
print((lr_gcv_model.best_score_)*100)
```

```python
final_lr = lr_gcv_model
```

```python
y_lr_final =  final_lr.predict(x_test_scaled)
```

```python
print(accuracy_score(y_lr_final , y_test)*100)
```

```python
print(classification_report(y_lr_final , y_test))
```

#### Clearly we can observe that optimised lr model (0.80,0.59)  has slightly less f1 score than the regular normal logisticregression (0.82 ,0.63) so general lf model is best
